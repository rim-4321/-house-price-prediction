```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 2: Create Dummy Dataset for Training
data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Bathrooms': [1, 2, 2, 3, 3],
    'Price': [100000, 150000, 200000, 250000, 300000]
}

df = pd.DataFrame(data)

# Step 3: Features & Target
X = df[['Area', 'Bedrooms', 'Bathrooms']]
y = df['Price']

# Step 4: Train Model
model = LinearRegression()
model.fit(X, y)

# Step 5: Take Manual Input from User
print("Enter details to predict house price:")

area = float(input("Enter area (in sqft): "))
bedrooms = int(input("Enter number of bedrooms: "))
bathrooms = int(input("Enter number of bathrooms: "))

# Step 6: Predict Price
input_data = pd.DataFrame([[area, bedrooms, bathrooms]], columns=['Area', 'Bedrooms', 'Bathrooms'])
predicted_price = model.predict(input_data)

# Step 7: Show Output
print(f"\nüè† Estimated House Price: ${predicted_price[0]:,.2f}")
```
